```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

a = np.arange(0., 40., 1)
f=a**2
g=1000*np.e**((-a)/25)
plt.plot(a,f)
plt.plot(a,g)
plt.ylabel("f(a) and g(a)")
plt.xlabel("a value")
blue_patch = mpatches.Patch(color='blue', label='f(a)')
orange_patch = mpatches.Patch(color='orange', label='g(a)')
plt.legend(handles=[blue_patch, orange_patch])
plt.show()
```

```python
import matplotlib.patches as mpatches
import numpy as np
import matplotlib.pyplot as plt

a = np.arange(0., 40., 1)
f=a**2
g=1000*np.e**((-a)/25)
p=f-g
plt.plot(a,p)
plt.ylabel("p(a)")
plt.xlabel("a-value")
plt.axhline(0, color='black')
blue_patch = mpatches.Patch(color='blue', label='p(a)')
plt.legend(handles=[blue_patch])
plt.show()
```
