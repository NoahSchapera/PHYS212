3. Write a function called *calder.py* whose inputs are the vector of lengths (assuming the same geometry as in the schematic above) and the total mass. Your function should output the individual mass values in this system.

```python
import numpy as np
def calder(rvec,tmass):
  #initialize array to store final massses in
  masses=np.zeros(8)
  counter = 0
  #initialize matrix of coefficients
  mat=[\
[rvec[0], -rvec[1], 0, 0, 0, 0, 0, 0],\
[0, rvec[2], -rvec[3], 0, 0, 0, 0, 0],\
[0, 0, 0, 0, rvec[4], -rvec[5], 0, 0],\
[0, 0, 0, 0, 0, 0, rvec[6], -rvec[7]],\
[rvec[8], rvec[8], -rvec[9], -rvec[9], 0, 0, 0, 0],\
[0, 0, 0, 0, rvec[10], rvec[10], -rvec[11], -rvec[11]],\
[rvec[12], rvec[12], rvec[12], rvec[12], -rvec[13], -rvec[13], -rvec[13], -rvec[13]],\
[1, 1, 1, 1, 1, 1, 1, 1]]
  #take inverse of coefficient matrix
  matinv=np.linalg.inv(mat)
  #make bvec vertical instead of horizontal
  bvecvert = np.vstack(bvec)
  #multiply inverse coefficients and bvec to find solution to system of equation 
  soln = np.matmul(matinv,bvecvert)
  #transpose solution matrix so its horizontal again
  soln=np.transpose(soln)
  #solutions are in a list of a list, for convenience convert to a single list
  for x in soln[0]:
    masses[counter]=x
    counter+=1
  return masses
  
```

4. Now let's say that we want to minimize the maximum value of any single mass in the system, and that to do this, we have the freedom to adjust $r_{13}$ anywhere in the range of $1$ cm to $50$ cm.

(a) Write a for loop containing your *calder.py* function that iteratively changes the value for $r_{13}$ and records the maximum mass in the system. Be sure to initialize any arrays that you will update within your for loop.

```python
#initialize rvec, r13 can be any value
rvec=[8,6,6,12,4,8,10,12,8,10,12,16,20,0]
#initialize matrix to sotre maximum masses in, 50 spots
maxMasses=np.zeros(50)
#iterate 50 times
for i in range(50):
  #reset the running maximum for each iteration, and increase the current r13 value by 1
  runningMax=0
  rvec[13]=i+1
  #find the list of masses for current r13 value
  m=calder(rvec,22869)
  #find maximum mass in this list
  for x in m:
    if runningMax < x:
      runningMax=x
  #set current maximum mass list index to the running maximum mass
  maxMasses[i]=runningMax
```
