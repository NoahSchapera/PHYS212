In this assignment, you'll continue to explore the behavior of systems with stochasticity that exhibit phase transitions.

In stochastic systems, the outcome of every simulation is different (due to the presence of random fluctuations). However, we expect that the system will reach equilibrium and therefore become *ergodic*, i.e., all accessible configurations are equiprobable over a long period of time. It follows from this that we can either run a single simulation for a very long time, or equivalently, we can average the behavior of multiple simulations with the same parameters. The latter is often less time consuming, so it can be beneficial to run multiple instantiations and average the results together. 

There are several hallmarks of a phase transition, like that seen in the Ising model. Depending on the nature of the phase transition (e.g., first order, second order), there are physical quantities in the system that can become undefined at the transition. One such quantity for the *Ising model* and for *percolation* (which you'll explore in the second question) is related to the sizes and variation of sizes of the clusters of spins.

2. *Node percolation*: Another relatively simple system that exhibits diverse behavior is percolation. In this system, each node is assumed to be "open" with probability p and "closed" with probability 1-p, and liquid can flow through open parts of the system. This model was inspired by geophysical flows (e.g., oil) through porous substrates, and developed to address the question: at what point can the liquid flow from one end of the system to the other?

(a) Given the description above, what is required for the liquid to flow from one side of the system to the other? How might we measure / quantify this?  

In order for liquid to flow, there there must be a path of adjacent nodes that are all open and connected to eachother from one side of the system to the other.  

We could quantidy this by identifying the largest cluster of connected 'open' pixels. If one end of this cluster is at the left side of the box, and the other end is at the opposite side, we know there must be an open path between these two points. Therefore, the system is percolatable.

We can find the coordinates of the largest clusters bounding box using the .bbox argument of regionprops. 

```python
import numpy.random as nprnd
```

(b) Write a function that creates an NxN matrix of random numbers between zero and 1, and binarizes this matrix by setting values smaller than a specified input threshold equal to 1 (these are open nodes) and everything else equal to zero (these are closed nodes). Your function should have inputs of N and threshold, and should output an NxN binary matrix. Note: this is simpler than the problem in class -- you are not being asked to compute/estimate a "best" threshold for the conversion to a binary image; instead, here, you'll vary this threshold value to explore how the system behaves.

```python
def porousMatrix(N,thresh):
  matrix=nprnd.rand(N,N)
  boolMat=matrix<thresh
  return boolMat
```

(c) Write a function that characterizes the open clusters in the binary image created by your function in part(b). Your function should return the label image,  number of clusters, the maximum cluster area, and the measures you proposed in part (a). Test your function on a small system (say N=100) for p=0.1 and p=0.9. Plot label images and discuss what you see. Discuss the results in the context of the qualitative picture for the model: what do these plots suggest about the flow of liquid through the system and why? 

```python
def porousProps(binary):
  percolatable=False
  #print(np.mean(binary))
  
  label_image = skimage.measure.label(binary)

  props = skimage.measure.regionprops(label_image)

  areas=np.array([p.area for p in props])
  numOfClusters=np.size(props)
  #print(areas.size)
  if areas.size==0:
    maxArea=0
  else:
    maxArea=np.amax(areas)
  
    #find region with the maximum area
    id = np.where(areas==maxArea)
    #find bounding box of maximum area region
    maxBbox=props[id[0][0]].bbox

    #determine if maximum bounding box stretches from one end of the region to the other
    if maxBbox[0]==0 and maxBbox[2]==binary.shape[0]:
      percolatable=True

  return label_image, numOfClusters, maxArea, percolatable
```

```python
mat01 = porousMatrix(100, 0.1)
mat09 = porousMatrix(100, 0.9)

label_image01, numOfClusters01, maxArea01, percolatable01 = porousProps(mat01)
label_image09, numOfClusters09, maxArea09, percolatable09 = porousProps(mat09)
```

```python
fig = plt.figure(figsize=(10,10))
ax=plt.subplot()
ax.set(title='System for P=0.1')
ax.imshow(label_image01)
```

```python
fig = plt.figure(figsize=(10,10))
ax=plt.subplot()
ax.set(title='System for P=0.9')
ax.imshow(label_image09)
```

When P=0.1, the system has a few disconnected open nodes in an otherwise closed matrix. The maximum number of connected open nodes in a cluster is only ~3-4. As a result, the system is not porous.

When P=0.9, the majority of nodes are open. As a result, the situation flips and the matrix is highly porous. There are no locations where the closed nodes completely block the path of open nodes from the left to the right. 

(d) To explore behavior of the model at intermediate values, create a finely and uniformly spaced array of probabilities, p (between 0 and 1, with 100 data points), and run your program above for each value of p and a system size of N = 1000. ```python

```python
probs=np.linspace(0,1,100)
N=1000

binaries=np.zeros((N,N,np.size(probs)))

labels=np.zeros((N,N,100))
clusterNums=np.zeros(100)
areas=np.zeros(100)
percs=np.zeros(100)


probMatrix=np.zeros((N,N,np.size(probs)))

for i in range(np.size(probs)):

  probMatrix[:,:,i] = porousMatrix(N,probs[i])
  
for i in range(np.size(probs)):
  labels[:,:,i],clusterNums[i],areas[i],percs[i]=porousProps(probMatrix[:,:,i])
```

```python
fig,ax = plt.subplots(1,1)
ax.plot(probs,clusterNums)
ax.set(xlabel='Probability', ylabel='Number of Clusters')
plt.show()
```

At P=0, there are no open nodes in the system and the number of clusters is zero. From P=0 to P=0.2, the number of nodes increases and the number of clusters of these open nodes also increases. However, at P=0.2 the probability of clusters connecting to one another begins to increase and the overall number of open clusters stars decreasing again. 

```python
fig,ax = plt.subplots(1,1)
ax.plot(probs,areas)
ax.set(xlabel='Probability', ylabel='Max Cluster Area')
plt.show()
```

At small values of P, we primarily see single open node 'clusters' and the maximum cluster size is very small. At P=0.4, we suddenly see a large spike in maximum cluster area. This point represents the transition between porous and nonporous matrices, as each cluster begins to connect to one another in an unbroken chain. The maximum cluster size continues to grow with probability until P=1, where every point is open and the cluster size is the size of the matrix. 

```python
fig,ax = plt.subplots(1,1)
ax.plot(probs,percs)
ax.set(xlabel='Probability', ylabel='Percolatablitiy (0=No, 1=Yes)')
plt.show()
```

At P=0, the cluster is nonporous as there are no open pores. At P=1, the cluster is completely open and therefore porous. As clusters begin to connect to eachother in larger and larger chains, the likelyhood of a matrix being porous increases. P=0.4 appears to represent the critical point where clusters are able to form chains the width of the matrix.
